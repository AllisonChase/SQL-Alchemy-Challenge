```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(autoload_with=engine)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table

Measurements = Base.classes.measurement
Stations = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
session.query(Measurements.date).order_by(Measurements.date.desc()).first()
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. Starting from the most recent data point in the database. 


# Calculate the date one year from the last date in data set.
one_year = dt.date(2017, 8, 23) - dt.timedelta(days=365)

# Perform a query to retrieve the data and precipitation scores
one_year_precipitation = session.query(Measurements.date, Measurements.prcp).filter(Measurements.date >= one_year).all()

# Save the query results as a Pandas DataFrame. Explicitly set the column names
one_year_precipitation = pd.DataFrame(one_year_precipitation, columns=['date', 'prcp'])

# Sort the dataframe by date
sorted_one_year_precipitation = one_year_precipitation.sort_values(by='date', ascending=False)

# Use Pandas Plotting with Matplotlib to plot the data
variable = one_year_precipitation.plot(x='date', y='prcp', figsize=(20,8), rot=75, color='blue')

plt.title("Precipitation in Hawaii over 12 Months", fontsize=30, weight='bold')
plt.xlabel("Date")
plt.ylabel("Precipitation (By Inches)")

plt.show()
```

```python
# Use Pandas to calculate the summary statistics for the precipitation data
one_year_precipitation.describe()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number of stations in the dataset
station_total = session.query(Stations.station).count()
station_total
```

```python
# Design a query to find the most active stations (i.e. which stations have the most rows?)
# List the stations and their counts in descending order.
#used Xpert AI to correct errors

active_stations = session.query(Measurements.station, func.count(Measurements.station)).\
    group_by(Measurements.station).\
    order_by(func.count(Measurements.station).desc()).all()

active_stations
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
first_station = active_stations[0][0]

session.query(func.min(Measurements.tobs), func.max(Measurements.tobs), func.avg(Measurements.tobs)).filter(Measurements.station==first_station).all()
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
#code help recieved from UCF tutor Justin

first_station_one_year = session.query(Measurements.tobs).filter(Measurements.date >= one_year).filter(Measurements.station==first_station).all()

plt.hist([x[0] for x in first_station_one_year], bins=12, color='blue')

plt.xlabel('Bins')
plt.ylabel('TOBS')
plt.title('Histogram')


plt.show() 
```

```python
# Close Session
session.close()
```
